```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

num_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])
cat_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("cat_encoder", OneHotEncoder(handle_unknown='ignore')),
    ])
from sklearn.compose import ColumnTransformer

num_attribs = numerical_columns
num_attribs2 = numerical_columns_test
cat_attribs = categorical_columns
cat_attribs2 = categorical_columns_test


preprocess_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", cat_pipeline, cat_attribs),        
    ])
preprocess_pipeline2 = ColumnTransformer([
        ("num", num_pipeline, num_attribs2),
        ("cat", cat_pipeline, cat_attribs2),        
    ])
```

```python
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)
```

```python
from sklearn.svm import LinearSVR

svm_clf = LinearSVR(epsilon=1.5, max_iter=5000)
svm_clf.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVR

rmse_scorer = make_scorer(rmse, greater_is_better=False)


# Perform cross-validation and calculate RMSE for each fold
rmse_scores = cross_val_score(svm_clf, x_train, y_train, cv=10, scoring=rmse_scorer)

# Output the mean RMSE
svr_mean_rmse = -rmse_scores.mean()
print("Mean RMSE:", svr_mean_rmse)
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.svm import LinearSVR
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_val_score

# Instantiate regressors
svm_clf = LinearSVR()
xgb = XGBRegressor()
lin_reg = LinearRegression()
model = LGBMRegressor()

# Create a voting regressor
voting_reg = VotingRegressor(
    estimators=[('svc', svm_clf), ('xgb', xgb), ('model', model), ('LinReg', lin_reg)]
)

# Define the custom RMSE scorer function
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)

# Perform cross-validation and calculate RMSE for each model
for clf in (svm_clf, xgb, model, lin_reg, voting_reg):
    rmse_scores = cross_val_score(clf, x_train, y_train, cv=10, scoring=rmse_scorer)
    mean_rmse = -rmse_scores.mean()
    print(f"{clf.__class__.__name__} Mean RMSE: {mean_rmse}")

# Fit the voting regressor
voting_reg.fit(x_train, y_train)
                                                                                                                                                                                                                                                                                                                                                                                                                        
```

```python
# Perform cross-validation and calculate RMSE for each fold
rmse_scores = cross_val_score(voting_reg, x_train, y_train, cv=10, scoring=rmse_scorer)

# Output the mean RMSE
ln_mean_rmse = -rmse_scores.mean()
print("Mean RMSE:", ln_mean_rmse)
```
